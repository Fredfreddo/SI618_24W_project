## Overview

By analyzing the data of more than 20,000 board games, we aim to analyze these games, evaluate what characteristics make them popular and highly rated, and fit models for popularity and rating base on statistical and machine learning techniques.

## Motivation

Board games have a long history, and are still evolving and immensely popular among modern players [**<font color = red>1</font>**](#reference). Many players have rated board games on online platforms, which give us access to measure how popular and appreciated by the players. Our motivation stems from the desire to delve deeper into the factors that contribute to the success and popularity of board games. By analyzing data from over 20,000 board games, we aim to unlock insights into what makes certain games stand out among the rest. Specifically, we aim to explore the following questions:
- What characteristics a board game has?
- What are the measurements for a board game's popularity and how well it is appreciated by its players?
- What factors, like playing time, minimum age, and number of players, affect the popularity of board games?
- What factors, like category and theme, affect the rating of board games?
- Are popular board games more highly rated, and vise versa?

## Data Source

The data files we use come from [kaggle](https://www.kaggle.com/datasets/joebeachcapital/board-games) [**<font color = red>2</font>**](#reference). The two datasets, `ratings.csv` and `details.csv`, according to kaggle page description, are acquired from [boardgamegeek.com](https://www.boardgamegeek.com/), which is an online forum for board gaming hobbyists and a game database that holds reviews, images and videos for over 125,600 different tabletop games, including European-style board games, wargames, and card games [**<font color = red>3</font>**](#reference).

`ratings.csv` contains the ratings information gathered from boardgamegeer.com of each board game, and `details.csv` contains the detailed information, such as publish year, number of players, and design information, such as designer and artist, of the board games. By combining these two datasets, we will be able to study the relationship between the various factors of the board games and their ratings and popularity.

#### Columns explanations:

- bayes_average: is a **Bayesian average** of the ratings [**<font color = red>4</font>**](#reference). It is calculated from the sample average from the online platform ratings and some prior mean and constant. We do not have information on the prior information the dataset used.

New merge by 'inner join'

```python
inner_data = pd.merge(ratings, details, on='id', how='inner')
```

```python
inner_data.isnull().sum()[inner_data.isnull().any()]
```

We noticed that the columns \[**boardgamecategory, boardgamemechanic, boardgamefamily, boardgameexpansion, boardgameimplementation, boardgamedesigner, boardgameartist, boardgamepublisher**\] are strings in the format of python lists, so we created new columns by converting them into list (while null values are converted into empty lists or a list with single element representing unknown).

```python
def string_to_list(string, default=None):
    """
    Converts a string representation of a list to a list.
    string: a string with a list representation, e.g. "[1, 2, 3]"
    default: the default value to return if the string is null
    """
    # if the string is null, return an empty list
    if pd.isnull(string):
        if default is not None:
            return [default]
        return []
    return eval(string)

# apply the function to the columns
inner_data['category_list'] = inner_data['boardgamecategory'].apply(string_to_list)
inner_data['mechanic_list'] = inner_data['boardgamemechanic'].apply(string_to_list)
inner_data['family_list'] = inner_data['boardgamefamily'].apply(string_to_list)
inner_data['expansion_list'] = inner_data['boardgameexpansion'].apply(string_to_list)
inner_data['implementation_list'] = inner_data['boardgameimplementation'].apply(string_to_list)
inner_data['artist_list'] = inner_data['boardgameartist'].apply(string_to_list, default='(Uncredited)')
inner_data['designer_list'] = inner_data['boardgamedesigner'].apply(string_to_list, default='(Uncredited)')
inner_data['publisher_list'] = inner_data['boardgamepublisher'].apply(string_to_list, default='Unknown')
```

```python
# deleting the original columns
inner_data.drop(['boardgamecategory', 'boardgamemechanic', 'boardgamefamily', 'boardgameexpansion', 'boardgameimplementation', 'boardgameartist', 'boardgamedesigner', 'boardgamepublisher'], axis=1, inplace=True)
inner_data.isnull().sum()[inner_data.isnull().any()]
```

We then decided to replace the null values in **thumbnail** and **description** with ' '

```python
# replace null with space character
inner_data['description'] = inner_data['description'].fillna(' ')
inner_data['thumbnail'] = inner_data['thumbnail'].fillna(' ')

# create a new DataFrame with the columns of interest, check for null values
ready_data = inner_data.copy()
ready_data.isnull().sum()[ready_data.isnull().any()]
```

We add columns recording number of category/family/mechanic/expansion/implementation/designer/publisher/artist a game has

```python
# add columns recording number of category/family/mechanic/expansion/implementation/designer/publisher/artist a game has
ready_data['category_count'] = ready_data['category_list'].apply(len)
ready_data['family_count'] = ready_data['family_list'].apply(len)
ready_data['mechanic_count'] = ready_data['mechanic_list'].apply(len)
ready_data['expansion_count'] = ready_data['expansion_list'].apply(len)
ready_data['implementation_count'] = ready_data['implementation_list'].apply(len)
ready_data['designer_count'] = ready_data['designer_list'].apply(len)
ready_data['publisher_count'] = ready_data['publisher_list'].apply(len)
ready_data['artist_count'] = ready_data['artist_list'].apply(len)
```
